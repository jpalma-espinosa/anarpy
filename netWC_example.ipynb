```python
import matplotlib.pyplot as plt
from anarpy.models import netWilsonCowan as nWC
from anarpy.models import netWilsonCowanPlastic as nWCp
```

```python
nWC.create("./anarpy/model_data/netWC.toml")
nWCp.create("./anarpy/model_data/netWC_plastic.toml")
```

```python
Vtrace,time=nWCp.Sim(verbose=True)
```

```python
plt.figure(1)
plt.clf()
plt.subplot(211)
plt.plot(time,Vtrace[:,0,3], color='C0', label='Excitatory')
plt.plot(time,Vtrace[:,1,3], color='C3', label='Inhibitory')
plt.xlim(0,1)
plt.legend()
```
